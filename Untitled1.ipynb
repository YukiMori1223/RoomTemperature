```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
data = pd.read_csv('data_c.csv',encoding = 'UTF8')

Tokyo=data.iloc[:,0]
Tokyo_min=data.iloc[:,1]
Tokyo_max=data.iloc[:,2]
Room=data.iloc[:,3]
Room_min=data.iloc[:,4]
Room_max=data.iloc[:,5]

x = pd.date_range('2020-10-01 00:00:00', periods=366, freq='d')

fig = plt.figure(figsize=(20,10))
plt.rcParams["font.size"] = 18

ax = plt.subplot(111)

ax.plot(x, Tokyo, label="Tokyo")
ax.fill_between(x, Tokyo_max, Tokyo_min, alpha=0.3)

ax.plot(x, Room, label="My Room")
ax.fill_between(x, Room_max, Room_min, alpha=0.3)

day = mdates.DayLocator() 
daysFmt = mdates.DateFormatter('%Y-%m')
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax.xaxis.set_major_formatter(daysFmt) 

#
sxmin='2020-10-01'
sxmax='2021-10-01'
xmin = datetime.datetime.strptime(sxmin, '%Y-%m-%d')
xmax = datetime.datetime.strptime(sxmax, '%Y-%m-%d')
plt.xlim([xmin,xmax])
ax.set_ylim(-5, 35)

#
ax.minorticks_on() #
ax.tick_params(axis="y", which="major",direction="in",length=5,width=2,top="on",right="on")
ax.tick_params(axis="y", which="minor",direction="in",length=2,width=1,top="on",right="on")

#
ax.set_xlabel("Date")
ax.set_ylabel("Temperature []")

plt.legend(bbox_to_anchor=(1, 0), loc='lower right', borderaxespad=1, fontsize=18)

ax.grid()

plt.show()
fig.savefig("img.png")
```
